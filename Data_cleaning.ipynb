```python
from langdetect import detect
import re
```

```python
df = pd.read_csv('flights.csv')
print(df.shape)
df
```

```python
df.info()
```

```python
df.dropna(subset=['rating'], inplace=True)
df.dropna(subset=['review'], inplace=True)
df.dropna(subset=['Route'], inplace=True)
```

```python
df.info()
```

```python
# Function to detect language
def detect_language(text):
    try:
        lang = detect(text)
    except:
        lang = 'unknown'
    return lang


# Filter rows by language (keeping only English)

for i in range(10):
    df = df[df['review'].apply(detect_language) == 'en']

df
```

```python
df.drop(columns=['Aircraft', 'Wifi & Connectivity'], inplace=True)

df.info()
    
```

```python
# Remove ordinal indicators
df['date'] = df['date'].str.replace(r'\b(\d+)(st|nd|rd|th)\b', r'\1', regex=True)

# Convert string dates to datetime in mm-dd-yyyy format
df['date'] = pd.to_datetime(df['date'], format='%d %B %Y').dt.strftime('%m-%d-%Y')

df
```

```python
date = pd.to_datetime(df['date'], errors='coerce')

df['Date Flown'] = df['Date Flown'].fillna(date.dt.strftime('%B %Y'))
```

```python
df.info()
```

```python
df['Verified'] = 0 
for index, row in df.iterrows():
    if 'Not Verified' in df['review'][index]:
        continue
        
    elif 'Verified' in df['review'][index]:
        df['Verified'][index] = 1
```

```python
df['review'] = df['review'].str.replace('Not Verified |', '')
df['review'] = df['review'].str.replace('✅ Trip Verified |', '')
```

```python
df
```

```python
df.info()
```

```python
median = df['Food & Beverages'].median()
df['Food & Beverages'].fillna(median, inplace=True)

mode = df['Type Of Traveller'].value_counts().idxmax()
df['Type Of Traveller'].fillna(mode, inplace=True)

median = df['Seat Comfort'].median()
df['Seat Comfort'].fillna(median, inplace=True)

median = df['Cabin Staff Service'].median()
df['Cabin Staff Service'].fillna(median, inplace=True)

median = df['Inflight Entertainment'].median()
df['Inflight Entertainment'].fillna(median, inplace=True)

median = df['Ground Service'].median()
df['Ground Service'].fillna(median, inplace=True)

median = df['Value For Money'].median()
df['Value For Money'].fillna(median, inplace=True)
```

```python
df.info()
```

```python
df['review_length'] = df['review'].apply(lambda x: len(x.split()))
```
