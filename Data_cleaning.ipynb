```python
df = pd.read_csv('enriched_data.csv')
df
```

```python
df.info()
```

```python
# Columns ‘Aircraft’, ‘Wifi & Connectivity’ and 'Route' are dropped as they contain many missing values.

df.drop(columns=['Aircraft', 'Wifi & Connectivity', 'Route'], inplace=True)
df.info()
```

# Keeping reviews only which are written in English language

```python
# Function to detect language
def detect_language(text):
    try:
        lang = detect(text)
    except:
        lang = 'unknown'
    return lang


# Filter rows by language (keeping only English)
for i in range(10):
    df = df[df['review'].apply(detect_language) == 'en']

df
```

# Converting 'date' to proper mm-dd-yyyy format

```python
# Remove ordinal indicators
df['date'] = df['date'].str.replace(r'\b(\d+)(st|nd|rd|th)\b', r'\1', regex=True)

# Convert string dates to datetime in mm-dd-yyyy format
df['date'] = pd.to_datetime(df['date'], format='%d %B %Y').dt.strftime('%m-%d-%Y')

df
```

```python
# Fill the missing 'Date Flown' value from its corresponding 'date' value 

date = pd.to_datetime(df['date'], errors='coerce')
df['Date Flown'] = df['Date Flown'].fillna(date.dt.strftime('%B %Y'))
```

# Extract verified status from the text of review

```python
df['Verified'] = 0
for index, row in df.iterrows():
    if 'Not Verified' in row['review']:
        continue
    elif 'Verified' in row['review']:
        df.loc[index, 'Verified'] = 1
```

```python
# Remove instances of verified status from the text of review
df['review'] = df['review'].str.replace('Not Verified |', '')
df['review'] = df['review'].str.replace('✅ Trip Verified |', '')
```

```python
df
```

# Imputation of missing values

```python
# Missing values for ‘Type Of Traveller’ are replaced with its mode.
mode = df['Type Of Traveller'].value_counts().idxmax()
df['Type Of Traveller'].fillna(mode, inplace=True)


# Missing values for following columns are replaced with their respective median values.
median = df['Food & Beverages'].median()
df['Food & Beverages'].fillna(median, inplace=True)

median = df['rating'].median()
df['rating'].fillna(median, inplace=True)

median = df['Seat Comfort'].median()
df['Seat Comfort'].fillna(median, inplace=True)

median = df['Cabin Staff Service'].median()
df['Cabin Staff Service'].fillna(median, inplace=True)

median = df['Inflight Entertainment'].median()
df['Inflight Entertainment'].fillna(median, inplace=True)

median = df['Ground Service'].median()
df['Ground Service'].fillna(median, inplace=True)

median = df['Value For Money'].median()
df['Value For Money'].fillna(median, inplace=True)
```

```python
df.info()
```

```python
# Add new column 'review_length' which specifies the word count of each review
df['review_length'] = df['review'].apply(lambda x: len(x.split()))
```

```python
# Encode the 'Recommended' column in integer format
df['Recommended'] = df['Recommended'].map({'yes': 1, 'no': 0})
```
