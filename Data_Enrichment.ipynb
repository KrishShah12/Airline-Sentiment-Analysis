# Generating country code based on country name

```python
import pycountry 


country_codes_2 = {
    'Czech Republic': 'CZ',
    'Macedonia': 'MK',
    'Vietnam': 'VN',
    'Turkey': 'TR',
    'Laos': 'LA',
    'Brunei': 'BN',
    'British Virgin Islands': 'VG',
    'South Korea': 'KR',
    'Venezuela': 'VE',
    'Trinidad & Tobago': 'TT',
    'US Virgin Islands': 'VI',
    'Moldova': 'MD',
    'Taiwan': 'TW',
    'Bolivia': 'BO',
    'Tanzania': 'TZ',
    'Iran': 'IR',
    'Macau': 'MO'
}

country_codes_3 = {
    'Czech Republic': 'CZE',
    'Macedonia': 'MKD',
    'Vietnam': 'VNM',
    'Turkey': 'TUR',
    'Laos': 'LAO',
    'Brunei': 'BRN',
    'British Virgin Islands': 'VGB',
    'South Korea': 'KOR',
    'Venezuela': 'VEN',
    'Trinidad & Tobago': 'TTO',
    'US Virgin Islands': 'VIR',
    'Moldova': 'MDA',
    'Taiwan': 'TWN',
    'Bolivia': 'BOL',
    'Tanzania': 'TZA',
    'Iran': 'IRN',
    'Macau': 'MAC'
}


def alpha2code(column, country_codes_2):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_2
            CODE.append(code)
        except:
            CODE.append(country_codes_2[country])
    return CODE

def alpha3code(column, country_codes_3):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
        except:
            CODE.append(country_codes_3[country])
    return CODE

df['CODE2'] = alpha2code(df.country, country_codes_2)
df['CODE3'] = alpha3code(df.country, country_codes_3)
df.head()
```

# Using Nominatim to get location

```python
from geopy.geocoders import Nominatim

# Initialize geocoder
geolocator = Nominatim(user_agent="data_wrgll")

# Get unique country codes
codes = df['CODE2'].unique()

# Function to geocode a single country code
def geocode_country_code(code):
    location = geolocator.geocode(code)
    if location:
        return pd.Series({'CODE2': code, 'Latitude': location.latitude, 'Longitude': location.longitude})
    else:
        return pd.Series({'CODE2': code, 'Latitude': None, 'Longitude': None})

# Create DataFrame with latitude and longitude pairs for each country code
geocoded_data = pd.DataFrame([geocode_country_code(code) for code in codes])
geocoded_data.head()
```

```python
# Merge geocoded data with original DataFrame
df = pd.merge(df, geocoded_data, on='CODE2', how='left')
```

```python
df.to_csv('enriched_data.csv', index=False)
```
